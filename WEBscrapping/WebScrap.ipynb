```python
# requests() Make a request to a web page, and print the response text

import requests
```

```python
#import the url of website to be scrapped 

ola = 'https://olaelectric.com/'
```

```python
#call requests

page = requests.get(ola)
```

```python
#print status (should get multiples of 2)

print(page.status_code)
```

```python
print(page)
```

```python
#if you want to print html data

print(page.text)
```

```python
#import bs4
from bs4 import BeautifulSoup
```

```python
#parse the data(html content) to convert it into readable format

soup = BeautifulSoup(page.text,'html.parser')
```

```python
# prettify() method will turn a Beautiful Soup parse tree into a nicely formatted Unicode string

print(soup.prettify)
```

```python
print(soup.find('a'))
```

```python
print(soup.findAll('a'))
```
