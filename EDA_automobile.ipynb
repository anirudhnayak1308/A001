```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/NetzwerkPriyanka8/ML_April/main/Projects/ED_Analysis/automobile.csv')
```

```python
df.head() #prints top 5
```

```python
df.tail()
```

```python
df.describe()  #descriptive statistics
```

#BOX PLOT

```python
sns.boxplot(x='num-of-cylinders',y='price',data=df)
```

#Scatter plot

```python
plt.scatter(x='engine-size',y='price',data=df)
plt.xlabel('Engine-size')
plt.ylabel('Price')
plt.grid()
plt.show
```

```python
sns.heatmap(df.isnull())
```

```python
count, bin_edges = np.histogram(df['peak-rpm'])
df['peak-rpm'].plot(kind='hist', xticks=bin_edges)
plt.xlabel('value of peak rpm')
plt.ylabel('Number of cars')
plt.grid()
plt.show()
```

#Correlation

```python
c_matrix = df.corr()
sns.heatmap(c_matrix, annot=True)
plt.show()
```

```python
sns.regplot(x='engine-size',y='price',data=df)
```
